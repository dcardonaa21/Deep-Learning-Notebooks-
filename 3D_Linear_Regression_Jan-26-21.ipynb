```python
import tensorflow as tf
from tensorflow import keras
```

```python
data = arreglo_norm
labels = data_set
```

```python
#1. Feature engineering-> los puntos estn normalizados?, estn en la misma dimensin?, grficar
figura = plt.figure()
ax = plt.axes(projection = "3d")
ax.scatter3D(data[:, 0], data[:, 1], labels, c=labels);
```

```python
modelo = tf.keras.Sequential([
                              keras.layers.Dense(units=1, input_shape=[2,])
                              ])
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer="rmsprop", loss="mean_squared_error")
```

```python
modelo.fit(data, labels, epochs=2500)
```

```python
modelo.weights
```

```python
noisy_labels = np.random.randn(len(labels)) 
```

```python
#1. Feature engineering-> los puntos estn normalizados?, estn en la misma dimensin?, grficar
figura = plt.figure()
ax = plt.axes(projection = "3d")
ax.scatter3D(data[:, 0], data[:, 1], noisy_labels, c=labels)
```

```python
#2. Crear modelo  
modelo_noise = tf.keras.Sequential([
                              keras.layers.Dense(units=1, input_shape=[2,])
                              ])
```

```python
modelo_noise.summary()
```

```python
modelo_noise.compile(optimizer="rmsprop", loss="mean_squared_error")
```

```python
modelo_noise.fit(data, noisy_labels, epochs=2500)
```

```python
modelo_noise.weights
```
