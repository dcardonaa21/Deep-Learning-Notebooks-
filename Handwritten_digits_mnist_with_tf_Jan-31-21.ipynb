## Feature engineering:
Scaling the images values to the same magnitud order:

```
Xs = train_mnist.loc[:, 1:785].values
Xs.shape
```

```
Xs = Xs.reshape(20000, 28, 28)
Xs_new = Xs[:, :, :, np.newaxis]
Xs_new.shape
```

```
Xs_new.max()
```

```
Xs_new= Xs_new/255.0
```

```
Ys = train_mnist.loc[:, 0].values
np.unique(Ys)
```

```
Xs_test = test_mnist.loc[:, 1:785].values
print(Xs_test.shape)
Xs_test = Xs_test.reshape(10000, 28, 28)
Xs_test = Xs_test[:, :, :, np.newaxis]
print(Xs_test.shape)
Xs_test = Xs_test/255.0
print(Xs_test.max())
```

```
Ys_test = test_mnist.loc[:, 0].values
np.unique(Ys_test)
```

## Model construction and validation:

```
model = tf.keras.Sequential(
    [
      tf.keras.layers.Conv2D(64, (3 ,3), activation="relu", input_shape=(28, 28, 1)),
      tf.keras.layers.MaxPool2D(2, 2),
      tf.keras.layers.Conv2D(64, (3 ,3), activation="relu"),
      tf.keras.layers.MaxPool2D(2, 2),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dense(128, activation="relu"),
      tf.keras.layers.Dropout(rate=0.2),
      tf.keras.layers.Dense(10, activation="softmax")
    ]
)
```

```
model.summary()
```

```
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=['accuracy'])
```

```
model.fit(Xs_new, Ys, epochs=7,
          callbacks=[tf.keras.callbacks.TensorBoard(log_dir="experiments/m_128-10Dense_DO0.2_Relu-Soft_7E")],
          validation_data=(Xs_test, Ys_test),
          shuffle=True)
```
